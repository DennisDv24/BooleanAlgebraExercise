```
import numpy as np
import math
import matplotlib.pyplot as plt
#import tensorflow as tf

_and  = lambda x,y:   np.logical_and(x,y)
_3and = lambda x,y,z: _and(_and(x,y),z)
_or   = lambda x,y:   np.logical_or (x,y)
_3or  = lambda x,y,z: _or(_or(x,y),z)
_not  = lambda x  :   np.logical_not(x)
_xor  = lambda x,y:   np.logical_xor(x,y)
defineM = lambda d1,d2: np.zeros((d1,d2))
```

```
def grayTObinary(valueTOconvert):
    valueTOconvert = int(valueTOconvert, 2)
    mask = valueTOconvert
    while mask != 0:
        mask >>= 1
        valueTOconvert ^= mask
    print('Binary array',bin(valueTOconvert)[2:])
      
    
 
valueTOconvert = input('Gray array to convert:')
grayTObinary(valueTOconvert)
```
